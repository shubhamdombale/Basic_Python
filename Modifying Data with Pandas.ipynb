```python
import numpy as np
import pandas as pd
```

```python
age = np.random.choice([15,20,45,12,'10',15,'34',7,'missing'],50)
fico=np.random.choice(['100-15', '150-200', '200-250', '250-300'], 50)

City = np.random.choice(['Mumbai', 'Thane', 'Pune', 'Latur'], 50)
ID=np.arange(50)
rating=np.random.choice(['Excellent', 'Good', 'Bad', 'Patheetic'], 50)
balance=np.random.choice([10000,20000,30000,40000,np.nan,50000,60000],50)
children=np.random.randint(high=5, low=0,size=(50))
```

```python
mydata=pd.DataFrame({'ID':ID,'age':age,'fico':fico,'City':City,'rating':rating,'balance':balance,'children':children})
```
